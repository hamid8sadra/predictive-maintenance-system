```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv('../data/predictive_maintenance_dataset.csv')
```

# Basic info

```python
print("Dataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
```

# Summary statistics

```python
print("\nSummary Statistics:")
print(df.describe())
```

# Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

# Define the target column based on the dataset

```python
target_col = 'failure'  # Explicitly using 'failure' as seen in the dataset
print(f"\nUsing '{target_col}' as the target column.")
```

# Visualize target distribution (failure or no failure)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=target_col, data=df)
plt.title('Distribution of Equipment Failures')
plt.savefig('../data/failure_distribution.png')
plt.show()
```

# Correlation heatmap for numeric features

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlations')
plt.savefig('../data/correlation_heatmap.png')
plt.show()
```
